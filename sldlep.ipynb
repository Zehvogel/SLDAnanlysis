```python
import ROOT
```

```python
from sldlep_utils import makeFilter
```

```python
%jsroot on
```

```python
#d = ROOT.RDataFrame("hermTree", "/nfs/dust/ilc/user/reichenl/SLDPID_tree_zh1000k.root")
d = ROOT.RDataFrame("hermTree", "/afs/desy.de/group/flc/pool/reichenl/myMarlinProcessors/SLDPIDTree/PID/results/SLDPID_tree_zhtest.root")
d = d.Define("trueCosTheta", "cos(trueTheta)")
```

```python
makeFilter("isPandoraJetMu", ("pandoraPDG", "isIsoLep"), "abs(pandoraPDG[i]) == 13 && isIsoLep[i] == 0")
makeFilter("isTruePandoraJetMu", ("pandoraPDG", "truePDG", "isIsoLep"), "abs(pandoraPDG[i]) == 13 && abs(truePDG[i]) == 13 && isIsoLep[i] == 0")
```

```python
makeFilter("isPandoraJetE", ("pandoraPDG", "isIsoLep"), "abs(pandoraPDG[i]) == 11 && isIsoLep[i] == 0")
makeFilter("isTruePandoraJetE", ("pandoraPDG", "truePDG", "isIsoLep"), "abs(pandoraPDG[i]) == 11 && abs(truePDG[i]) == 11 && isIsoLep[i] == 0")
```

```python
purdata = (d.Define("isPandoraJetMu", "isPandoraJetMu(pandoraPDG, isIsoLep)")
            .Define("isTruePandoraJetMu", "isTruePandoraJetMu(pandoraPDG, truePDG, isIsoLep)")
            .Define("isPandoraJetE", "isPandoraJetE(pandoraPDG, isIsoLep)")
            .Define("isTruePandoraJetE", "isTruePandoraJetE(pandoraPDG, truePDG, isIsoLep)"))
```

```python
allPandoraJetMu = purdata.Histo2D(("allPandoraJetMu", "pandora #mu; cos#theta; Momentum (GeV)", 50, -1, 1, 50, 0, 50), "trueCosTheta", "trueP", "isPandoraJetMu")
truePandoraJetMu = purdata.Histo2D(("truePandoraJetMu", "pandora #mu true; cos#theta; Momentum (GeV)", 50, -1, 1, 50, 0, 50), "trueCosTheta", "trueP", "isTruePandoraJetMu")
```

```python
allPandoraJetE = purdata.Histo2D(("allPandoraJetE", "pandora e; cos#theta; Momentum (GeV)", 50, -1, 1, 50, 0, 50), "trueCosTheta", "trueP", "isPandoraJetE")
truePandoraJetE = purdata.Histo2D(("truePandoraJetE", "pandora e true; cos#theta; Momentum (GeV)", 50, -1, 1, 50, 0, 50), "trueCosTheta", "trueP", "isTruePandoraJetE")
allPandoraJetE10GeV = purdata.Histo2D(("allPandoraJetE", "pandora e; cos#theta; Momentum (GeV)", 50, -1, 1, 50, 0, 10), "trueCosTheta", "trueP", "isPandoraJetE")
truePandoraJetE10GeV = purdata.Histo2D(("truePandoraJetE", "pandora e true; cos#theta; Momentum (GeV)", 50, -1, 1, 50, 0, 10), "trueCosTheta", "trueP", "isTruePandoraJetE")
```

```python
pandoraJetMuPurity = truePandoraJetMu.GetPtr().Clone()
pandoraJetMuPurity.Divide(allPandoraJetMu.GetPtr())
```

```python
pandoraJetEPurity = truePandoraJetE.GetPtr().Clone()
pandoraJetEPurity.Divide(allPandoraJetE.GetPtr())
pandoraJetEPurity10GeV = truePandoraJetE10GeV.GetPtr().Clone()
pandoraJetEPurity10GeV.Divide(allPandoraJetE10GeV.GetPtr())
```

```python
c1 = ROOT.TCanvas()
pandoraJetMuPurity.SetTitle("pandora jet #mu pur; cos(#theta); Momentum (GeV)")
pandoraJetMuPurity.Draw("COLZ")
c1.Draw()
```

```python
c2 = ROOT.TCanvas()
pandoraJetEPurity.SetTitle("pandora jet e pur; cos(#theta); Momentum (GeV)")
pandoraJetEPurity.Draw("COLZ")
c2.Draw()
```

```python
c3 = ROOT.TCanvas()

truePandoraJetE.Draw("COLZ")

c3_2 = ROOT.TCanvas()
relTruePandoraJetE = truePandoraJetE.GetPtr().Clone()
norm = 1 / relTruePandoraJetE.Integral()
relTruePandoraJetE.Scale(norm)
relTruePandoraJetE_projP = relTruePandoraJetE.ProjectionY()
relTruePandoraJetE_projP_cum = relTruePandoraJetE_projP.GetCumulative()
relTruePandoraJetE_projP_cum.Draw("HIST")
c3.Draw()
c3_2.Draw()
```

```python
c4 = ROOT.TCanvas()
allPandoraJetE.Draw("COLZ")
c4_2 = ROOT.TCanvas()
relAllPandoraJetE = allPandoraJetE.GetPtr().Clone()
norm = 1 / relAllPandoraJetE.Integral()
relAllPandoraJetE.Scale(norm)
relAllPandoraJetE_projP = relAllPandoraJetE.ProjectionY()
relAllPandoraJetE_projP_cum = relAllPandoraJetE_projP.GetCumulative()
relAllPandoraJetE_projP_cum.Draw("HIST")
c4.Draw()
c4_2.Draw()
```
